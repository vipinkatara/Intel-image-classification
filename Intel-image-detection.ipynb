# Importing basic libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import cv2
import os
```

# Loading the train data

```python
DATA_DIR = './seg_train/seg_train'
CATEGORIES = ['buildings', 'forest', 'glacier', 'mountain', 'sea', 'street']
IMG_SIZE = 50
data = []

for category in CATEGORIES:
    path = os.path.join(DATA_DIR, category)
    class_num = CATEGORIES.index(category)
    
    for img in os.listdir(path):
        try:
            img_array = cv2.imread(os.path.join(path, img), 0)
            new_array = cv2.resize(img_array, (IMG_SIZE, IMG_SIZE))
            data.append([new_array, class_num])
        except Exception as e:
            pass
    
```

```python
np.shape(data)
```

# separating features and labels

```python
X = []
Y = []
for feature, labels in data:
    X.append(feature)
    Y.append(labels)
    
X = np.array(X).reshape(-1, IMG_SIZE, IMG_SIZE, 1)
```

# Normalization

```python
X = X / 255
```

# Converting to categorical values

```python
from keras.utils import to_categorical
Y = to_categorical(Y)
```

# Loading the test data

```python
DATA_DIR = './seg_test/seg_test'
CATEGORIES = ['buildings', 'forest', 'glacier', 'mountain', 'sea', 'street']
IMG_SIZE = 50
data = []

for category in CATEGORIES:
    path = os.path.join(DATA_DIR, category)
    class_num = CATEGORIES.index(category)
    
    for img in os.listdir(path):
        try:
            img_array = cv2.imread(os.path.join(path, img), 0)
            new_array = cv2.resize(img_array, (IMG_SIZE, IMG_SIZE))
            data.append([new_array, class_num])
        except Exception as e:
            pass
    
```

```python
np.shape(data)
```

# Separating features and labels of test data

```python
X_test = []
Y_test = []

for feature, label in data:
    X_test.append(feature)
    Y_test.append(label)

X_test = np.array(X_test).reshape(-1,IMG_SIZE,IMG_SIZE,1)
```

```python
#normalization
X_test = X_test / 255
```

# Spliting the dataset into train and val

```python
from sklearn.model_selection import train_test_split
x_train, x_val, y_train, y_val = train_test_split(X, Y, test_size=0.2, random_state=0)
```

# Data Augumentation

```python
from keras.preprocessing.image import ImageDataGenerator

datagen = ImageDataGenerator(

     featurewise_center=False,  # set input mean to 0 over the dataset
        samplewise_center=False,  # set each sample mean to 0
        featurewise_std_normalization=False,  # divide inputs by std of the dataset
        samplewise_std_normalization=False,  # divide each input by its std
        zca_whitening=False,  # apply ZCA whitening
        rotation_range=10,  # randomly rotate images in the range (degrees, 0 to 180)
        zoom_range = 0.1, # Randomly zoom image 
        width_shift_range=0.1,  # randomly shift images horizontally (fraction of total width)
        height_shift_range=0.1,  # randomly shift images vertically (fraction of total height)
        horizontal_flip=False,  # randomly flip images
        vertical_flip=False  # randomly flip images

    
)
```

# CNN

```python
from keras.models import Sequential
from keras.layers import Conv2D, MaxPool2D, Dense, Dropout, Flatten
import keras

model = Sequential()
model.add(Conv2D(64, 3, activation='relu',padding = 'same', data_format='channels_last', input_shape=(IMG_SIZE,IMG_SIZE,1)))
model.add(Conv2D(16, 3, activation='relu', padding='same'))
model.add(Conv2D(4, 3, activation='relu', padding='same'))
model.add(MaxPool2D(pool_size=(2,2)))
model.add(Flatten())
model.add(Dense(64))
model.add(Dropout(0.3))
model.add(Dense(16, activation='relu'))
model.add(Dropout(0.3))
model.add(Dense(6))
model.add(Dense(6, activation='softmax'))
model.compile(loss = keras.losses.categorical_crossentropy, optimizer=keras.optimizers.Adam(),
              metrics=['accuracy'])
model.summary()
```

```python
batch_size = 64
epochs = 50
history = model.fit_generator(datagen.flow(x_train,y_train, batch_size=batch_size),
                              epochs = epochs, validation_data = (x_val,y_val),
                              )
```

# Ploting training loss vs validation loss

```python
loss = history.history['loss']
val_loss = history.history['val_loss']
epochs = range(1, len(loss) + 1)
plt.plot(epochs, loss, 'y', label='Training loss')
plt.plot(epochs, val_loss, 'r', label='Validation loss')
plt.title('Training and validation loss')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.legend()
plt.show()
```

# Ploting training accuracy vs validation accuracy

```python
acc = history.history['accuracy']
val_acc = history.history['val_accuracy']
plt.plot(epochs, acc, 'y', label='Training acc')
plt.plot(epochs, val_acc, 'r', label='Validation acc')
plt.title('Training and validation accuracy')
plt.xlabel('Epochs')
plt.ylabel('Accuracy')
plt.legend()
plt.show()
```

# Predicting values

```python
y_pred = model.predict(X_test)
```

# converting labels of test dataset to categorical

```python
Y_test = to_categorical(Y_test)
```

# Checking accuracy of predictions

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_pred.round(), Y_test)
```
